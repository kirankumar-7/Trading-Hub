```python
from binance.client import Client
import pandas as pd

client = Client()

# Define date range and interval for historical data
start_str = '1 Jan, 2020'
end_str = '2 Jan, 2020'
interval = '1d'

def snapshot():
    # Open and read the symbols from the CSV file
    with open('datasets/crypto_usdt_symbols.csv', 'r') as f:
        for symbol in f:  
            try:
                # Fetch historical data from Binance
                data = client.get_historical_klines(symbol, interval, start_str, end_str)
                
                # Convert the data to a DataFrame with appropriate column headers
                df = pd.DataFrame(data, columns=[
                    'Open time', 'Open', 'High', 'Low', 'Close', 'Volume', 
                    'Close time', 'Quote asset volume', 'Number of trades', 
                    'Taker buy base asset volume', 'Taker buy quote asset volume', 'Ignore'
                ])
                
                # Save the DataFrame to a CSV file
                df.to_csv(f'datasets/crypto_usdt_daily_data/{symbol}.csv', index=False)
            except Exception as e:
                print(f"Failed to fetch data for {symbol}: {e}")
    
    return 'Data Extrated from binance API for the given symbols successfully'

snapshot()
```

```python
from binance.client import Client
import pandas as pd
from datetime import datetime

client = Client()

start_str = '1 Jan, 2020'
end_str = '2 Jan, 2020'
interval = '1d'
symbol = 'BTCUSDT'

data = client.get_historical_klines(symbol, interval, start_str, end_str)

processed_data = []
for x in data:
    row = [
        datetime.fromtimestamp(x[0] / 1000).strftime('%Y-%m-%d'),  # Convert timestamp to Date
        float(x[1]),  # Open
        float(x[2]),  # High
        float(x[3]),  # Low
        float(x[4]),  # Close
        float(x[5]),  # Volume
        int(x[8])     # Number of trades
    ]
    processed_data.append(row)

df = pd.DataFrame(processed_data, columns=[
    'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Number of trades'
])

print(df)
```
